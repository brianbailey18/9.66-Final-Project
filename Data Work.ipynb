```python
df_2022 = pd.read_csv(f'/content/drive/My Drive/MIT Year 4/Fall Semester/9.66/9.66 Final Project/NFL_games_by_week_2022.csv', index_col=0)
df_2022
```

```python
result = {'home_team' : [], 'away_team' : [], 'home_score' : [], 'away_score' : []}

for i, row in df_2022.iterrows():

  if row['At'] == '@':

    home_idx, away_idx = 'Loser/tie', 'Winner/tie'
    home_score_idx, away_score_idx = 'PtsL', 'PtsW'

  else: 
    
    home_idx, away_idx = 'Winner/tie', 'Loser/tie'
    home_score_idx, away_score_idx = 'PtsW', 'PtsL'


  home_team, away_team = row[home_idx], row[away_idx]
  home_score, away_score = row[home_score_idx], row[away_score_idx]

  result['home_team'].append(home_team)
  result['away_team'].append(away_team)
  result['home_score'].append(home_score)
  result['away_score'].append(away_score)

edited_df = pd.DataFrame.from_dict(result)
edited_df
```
